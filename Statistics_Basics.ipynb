```python
# What is the difference between descriptive statistics and inferential statistics? Explain with examples.
'''
Descriptive statistics:
→ Summarizes existing data.
Example: Average marks of a class = 72.

Inferential statistics:
→ Uses sample data to predict about a population.
Example: Estimating average marks of the whole school using one class data.
'''
```

```python
# What is sampling in statistics? Explain the differences between random and stratified sampling.
'''
Sampling: Selecting a small group (sample) from a population to study and draw conclusions.

Random Sampling: Every member has equal chance of selection.
Example: Randomly picking 50 students from a school.

Stratified Sampling: Population divided into groups (strata), then samples taken from each group.
Example: Selecting students from each class (Class 9, 10, 11, 12).
'''
```

```python
# Define mean, median, and mode. Explain why these measures of central tendency are important.
'''
Mean: Average of all values.
Example: (2, 4, 6)  mean = 4

Median: Middle value when data is ordered.
Example: (2, 4, 6)  median = 4

Mode: Most frequent value.
Example: (2, 4, 4, 6) mode = 4

Why they are important:
- Show the central/typical value of data
- Help summarize and compare datasets
- Useful for decision-making and analysis
'''
```

```python
# Explain skewness and kurtosis. What does a positive skew imply about the data?
'''
Skewness: Measures asymmetry of a data distribution.

Kurtosis: Measures peakedness / tail heaviness of a distribution.

Positive skew (right skew):
- Long tail on the right side
- Most values are low
- Mean > Median > Mode
'''
```

```python
# Implement a Python program to compute the mean, median, and mode of a given list of numbers.

import statistics
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
mean = statistics.mean(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
# Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:

import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Covariance matrix
cov_matrix = np.cov(list_x, list_y)
covariance = cov_matrix[0][1]

# Correlation coefficient
correlation = np.corrcoef(list_x, list_y)[0][1]

print("Covariance:", covariance)
print("Correlation coefficient:", correlation)
```

```python
# You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
# Explain how you would use covariance and correlation to explore this relationship.
# Write Python code to compute the correlation between the two lists:

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

correlation = np.corrcoef(advertising_spend, daily_sales)[0][1]
print("Correlation:", correlation)
```
