3. What are pre-defined keywords in Python?

5. What is mutability in Python?
    - In python mutability means when object is allowed to manipulate/change at place(without creating new object) like list, dict etc...
6. Why are lists mutable, but tuples are immutable?
    - Its in nature of lists to allow add, update or delete element which is usefull for dynamic manipulation. In contrast, tuples guarantee that their contents remain constant, which prevents accidental changes and improve performance because tuples are stored in one single memory block.
7. What is the difference between “==” and “is” operators in Python?
    - "==" checks if the value of two variable are same(equality) and "is" checks if two variables point to same object in memory(identity).
8. What are logical operators in Python?

11. What is the purpose of conditional statements in Python?
    - The purpose of conditional statements in python is to allow a block of code to run if certain condition is matched allowing the program to make a decision and control its flow.
12. How does the elif statement work?

```python
# Write a function to demonstrate the behavior of mutable and immutable arguments.

def modify_values(num, lst):
  num+=10
  lst.append(10)
  return num, lst

x = 5
my_list = [1,2,3]

print("Before: ", x, my_list)
new_x, new_list = modify_values(x, my_list)
print("After: ", new_x, new_list)
print("Original: ", x, my_list)
```

```python
# Write a program that performs basic arithmetic operations on two user-input numbers.
import ast

num1 = input("Enter first input: ")
num2 = input("Enter second input: ")
opt = input("Enter operation: ")

if opt in ['-', '+', '*', '/']:
  try:
    print(f"{num1} {opt} {num2} = {eval(f"{num1} {opt} {num2}")}")
  except Exception as e:
    print(e)
else:
  print("Invalid Operator")
```

```python
# Write a program to reverse a string using a while loop.
text = input("Enter String: ")
reversed_text = ''
i = len(text)-1
while i>=0:
  reversed_text += text[i]
  i -= 1
print("Reversed string: ", reversed_text)
```

```python
# Write a Python program to calculate the factorial of a number provided by the user using a while loop

num = int(input("Enter a number: "))
factorial = 1
i = 1

while i<= num:
  factorial *= i
  i+=1

print("Factorial of", num, "is", factorial)
```
