```python
# What is a parameter?
'''
Parameter
  A numerical value that describes a population
  Usually fixed but unknown

Example:
  True average income of all customers
  True population proportion
  Parameter = population value
'''
```

```python
# What is correlation?
'''
Measures the strength and direction of relationship between two variables
Value ranges from −1 to +1
'''
# What does negative correlation mean?
'''
As one variable increases, the other decreases
Example: Price high -> Demand low
'''
```

```python
# Define Machine Learning. What are the main components in Machine Learning?
'''
Machine Learning
  A method where computers learn from data to make predictions or decisions without being explicitly programmed.

Main components:
  Data – Input information
  Model – Algorithm that learns patterns
  Training – Process of learning from data
  Evaluation – Measuring performance
  Prediction – Output result
'''
```

```python
# How does loss value help in determining whether the model is good or not?
'''
Loss value
  Measures the error between predicted and actual values

How it helps:
  Lower loss → Better model performance
  Shows how well the model is learning
  Used to compare models and improve them
  Goal: Minimize loss during training
'''
```

```python
# What is causation? Explain difference between correlation and causation with an example.
'''
Causation
  When one variable directly causes change in another.

Correlation
  Two variables move together, but one may not cause the other.

Difference:
  Correlation = relationship
  Causation = cause-and-effect

Example:
Ice cream sales ↑ and drowning cases ↑ in summer.
They are correlated, but ice cream does not cause drowning.
The real cause is hot weather.
'''
```

```python
# What is an Optimizer? What are different types of optimizers? Explain each with an example.
'''
Optimizer
  An algorithm used to minimize the loss function by updating model weights.

Types of optimizers:
  Gradient Descent (GD)
    Uses entire dataset to update weights
    Stable but slow

  Stochastic Gradient Descent (SGD)
    Updates weights using one data point at a time
    Faster but noisy

  Mini-batch Gradient Descent
    Uses small batches of data
    Balanced speed & stability

  Adam (Adaptive Moment Estimation)
    Adjusts learning rate automatically
    Fast and widely used in deep learning
All optimizers aim to reduce error and improve model accuracy.
'''
```

```python
# What does model.fit() do? What arguments must be given?
'''
model.fit()
  Trains the model on data
  Learns patterns by adjusting weights

Arguments required:
  X → Input features (training data)
  y → Target/output values

Example: model.fit(X_train, y_train)
'''
```

```python
# What does model.predict() do? What arguments must be given?
'''
model.predict()
  Uses the trained model to make predictions
  Applies learned patterns to new data

Argument required:
  X → Input data (usually test data)
Example: model.predict(X_test)
'''
```

```python
# What is feature scaling? How does it help in Machine Learning?
'''
Feature scaling
  Process of bringing features to a similar scale

Why it helps:
  Improves model performance
  Speeds up training
  Prevents one feature from dominating others

Important for models like KNN, SVM, Linear Regression, Neural Networks.
'''
```
