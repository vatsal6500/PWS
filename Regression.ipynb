```python
# What is Simple Linear Regression?
'''
- A method to model a relationship between one independent variable(x) and one dependent variable(y).
- Fits a straight line to predict y
- y = a + bX
'''
```

```python
# What are the key assumptions of Simple Linear Regression?
'''
- Key assumptions of SLR:
  - Linearity – Relationship between X and Y is linear
  - Independence – Observations are independent
  - Homoscedasticity – Constant variance of errors
  - Normality – Errors are normally distributed
  - No outliers – Extreme values should not distort the model
'''
```

```python
# What does the coefficient m represent in the equation Y=mX+c?
'''
Represents the rate of change of Y with respect to X
Shows how much Y changes when X increases by 1 unit
'''
```

```python
# What does the intercept c represent in the equation Y=mX+c?
'''
The value of Y when X = 0
It is the point where the line crosses the Y-axis.
'''
```

```python
# How do we calculate the slope m in Simple Linear Regression?
'''
Measures change in Y for one unit change in X
Determines the steepness of the regression line.
'''
```

```python
# What is the purpose of the least squares method in Simple Linear Regression?
'''
Minimizes the sum of squared differences between actual and predicted values
'''
```

```python
# How is the coefficient of determination (R²) interpreted in Simple Linear Regression?
'''
Measures how well the regression model explains the variation in Y
'''
```

```python
# What is Multiple Linear Regression?
'''
A regression method that uses two or more independent variables (X₁, X₂, …) to predict one dependent variable (Y).
'''
```

```python
# What is the main difference between Simple and Multiple Linear Regression?
'''
Simple Linear Regression
  Uses one independent variable (X) to predict Y

Multiple Linear Regression
  Uses two or more independent variables (X1, X2, …) to predict Y
'''
```

```python
# What are the key assumptions of Multiple Linear Regression?
'''
- Key assumptions of MLR
  - Linearity - Linear relationship between predictors and target
  - Independence - Observations are independent
  - Homoscedasticity - Constant variance of errors
  - Normality - Residuals are normally distributed
  - No multicollinearity - Independent variables should not be highly correlated
'''
```

```python
# What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?
'''
Heteroscedasticity
  When the variance of errors is not constant across all values of the independent variables.

Effect on regression:
  Makes coefficient estimates less reliable
  Affects standard errors and hypothesis tests
  Can lead to incorrect conclusions about significance.
'''
```

```python
# How can you improve a Multiple Linear Regression model with high multicollinearity?
'''
Remove highly correlated variables
Use feature selection
Apply Ridge or Lasso regression
Use Principal Component Analysis (PCA)
'''
```

```python
# What are some common techniques for transforming categorical variables for use in regression models?
'''
One-Hot Encoding – Creates binary columns for each category
Label Encoding – Assigns numbers to categories
Ordinal Encoding – For ordered categories
'''
```

```python
# What is the role of interaction terms in Multiple Linear Regression?
'''
Interaction terms:
  Represent the combined effect of two variables on the target.

Role:
  Show how the effect of one variable changes depending on another.
'''
```

```python
# How can the interpretation of intercept differ between Simple and Multiple Linear Regression?
'''
Simple Linear Regression
  Intercept = value of Y when X = 0

Multiple Linear Regression
  Intercept = value of Y when all independent variables (X1, X2, …) = 0
'''
```

```python
# What is the significance of the slope in regression analysis, and how does it affect predictions?
'''
Slope (m)
  Shows the change in Y for a one-unit increase in X

Significance:
  Indicates the strength and direction of relationship
  Used to calculate predictions
'''
```

```python
# How does the intercept in a regression model provide context for the relationship between variables?
'''
Intercept
  The predicted value of Y when all independent variables are 0

How it provides context:
  Acts as the baseline value of Y
  Helps understand how Y changes as X increases from the starting point.
'''
```

```python
# What are the limitations of using R² as a sole measure of model performance?
'''
Does not show if the model is correct or unbiased
Always increases when more variables are added
Does not detect overfitting
Does not indicate causation
'''
```

```python
#  How would you interpret a large standard error for a regression coefficient?
'''
Large standard error (for a coefficient)
  Indicates the estimate is less precise

Interpretation:
  The coefficient may not be statistically significant
  The predictor may have weak or uncertain effect on Y.
'''
```

```python
# How can heteroscedasticity be identified in residual plots, and why is it important to address it?
'''
Identify in residual plot:
  Residuals show uneven spread
  Pattern like funnel or cone shape instead of random scatter

Why address it:
  Makes standard errors unreliable
  Affects hypothesis tests and confidence intervals
  Can lead to incorrect conclusions.
'''
```

```python
# What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?
'''
The model includes too many unnecessary predictors

Explanation:
  R² increases when more variables are added
  Adjusted R² decreases if variables do not improve the model
'''
```

```python
# Why is it important to scale variables in Multiple Linear Regression?
'''
Brings variables to same scale
Prevents features with large values from dominating the model
Improves model stability and interpretation.
'''
```

```python
# What is polynomial regression?
'''
Polynomial Regression
  A regression method that models non-linear relationships between variables
  Uses polynomial terms (X², X³, etc.)
'''
```

```python
# How does polynomial regression differ from linear regression?
'''
Linear Regression
  Models a straight-line relationship between X and Y.

Polynomial Regression
  Models a curved relationship using terms like X^2, X^3.
'''
```

```python
# When is polynomial regression used?
'''
Relationship between variables is non-linear
Data shows a curved pattern
Linear regression cannot fit the data well
'''
```

```python
# What is the general equation for polynomial regression?
'''
General equation for Polynomial Regression

Y = b0 + b1X + b2X2 + b3X3 +...+ bnXn

b₀ = intercept
b₁, b₂, ... = coefficients
n = degree of polynomial.
'''
```

```python
# Can polynomial regression be applied to multiple variables?
'''
Polynomial regression can be applied to multiple variables by adding polynomial terms for each variable.
'''
```

```python
# What are the limitations of polynomial regression?
'''
Can cause overfitting with high-degree polynomials
Model becomes complex and hard to interpret
Sensitive to outliers
Performs poorly for extrapolation outside data range
'''
```

```python
#  What methods can be used to evaluate model fit when selecting the degree of a polynomial?
'''
R2 / Adjusted R2 – Measure goodness of fit
Cross-validation – Check performance on unseen data
MSE / RMSE – Measure prediction error
Residual plots – Check if errors are random
'''
```

```python
# Why is visualization important in polynomial regression?
'''
Helps see non-linear patterns in data
Shows how well the curve fits the data
Helps detect underfitting or overfitting
'''
```
