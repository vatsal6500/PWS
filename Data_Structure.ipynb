1. What are data structures, and why are they important?
    - Data structures are ways to orginize data for efficient access and modification, they improve performance, memory efficiency and scalability of programs
1. Explain the difference between mutable and immutable data types with examples.
    - Mutable: can be changes after creation. eg: list, dict, set
    - Immutable cannot be changed after creation. eg: int, float, str, tuple
1. What are the main differences between lists and tuples in Python?
    - Lists are mutable while tuples are immutable, faster and memory efficient.
1. Describe how dictionaries store data.
    - Dict stores data in key-value pair where key is named index to its value.
1. Why might you use a set instead of a list in Python?
    - Might use set when uniqueness is required without handling explicitly.
1. What is a string in Python, and how is it different from a list?
    - String is sequence of characters enclosed in quotes, while list is a sequence of items that can include different data types.
1. How do tuples ensure data integrity in Python?
    - By being immutable meaning once they are created their elements cannot be changed, added or removed.
1. What is a hash table, and how does it relate to dictionaries in Python?
    - Hash table is a data stucture that stores key:value pairs, allowing for fast data retrieval using hash function to generate an index for each key.
1. Can lists contain different data types in Python?
    - Yes.
1. Explain why strings are immutable in Python.
    - Strings are immutable by design for following reasons improve performance, reliable hashing, memory optimization and sharing across programs.
1. What advantages do dictionaries offer over lists for certain tasks?
    - Dict offers access to value using key(named) not just index like list.
1. Describe a scenario where using a tuple would be preferable over a list.
    - When you need to store collection of items that should not change, such as coordination of point in a 2D space. eg: (x,y) = (23,45)
1. How do sets handle duplicate values in Python?
    - Set using hashing, so values can appear only once.
1. How does the “in” keyword work differently for lists and dictionaries?
    - In list, it checks if a value exists among the elements, while in dict, it checks if a key exists amoung the dict's key.
1. Can you modify the elements of a tuple? Explain why or why not.
    - No because tuples are immutable in python, if tuple contains mutable object then that object can be modified.
1. What is a nested dictionary, and give an example of its use case?
    - A dict that contains another dict as a value.
    - Used to represent structured or hierarchical data.
1. Describe the time complexity of accessing elements in a dictionary.
    - Dictionary elements access is O(1) on average due to hashing, but can degrade to O(n) in worst case.
1. In what situations are lists preferred over dictionaries?
    - Ordered, index based collections that allows duplicates and don't require key:value pair.
1. Why are dictionaries considered unordered, and how does that affect data retrieval?
    - Dict are based on hash table, so elements are stored by hash value, not by position and access is done by key not index.
1. Explain the difference between a list and a dictionary in terms of data retrieval.
    - In list data is reterived using a numeric value while in dict data is reterived using a key.
