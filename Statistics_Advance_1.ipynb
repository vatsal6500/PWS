```python
#  What is a random variable in probability theory?
'''
A function that assigns numerical values to outcomes of a random experiment.
'''
```

```python
# What are the types of random variables?
'''
Types of random variables:
- Discrete random variable
  - Takes countable values
  Example: Number of customers in a store

- Continuous random variable
  - Takes any value in a range
  Example: Height or time
'''
```

```python
# Explain the difference between discrete and continuous distributions
'''
Discrete distribution
  - Values are countable
  - Uses PMF
  Example: Number of customers

Continuous distribution
  - Values are in a range (uncountable)
  - Uses PDF
  Example: Time or height
'''
```

```python
# What is a binomial distribution, and how is it used in probability?
'''
Binomial distribution
  - A discrete probability distribution
  - Counts number of successes in fixed number of independent trials
  - Each trial has two outcomes (success/failure)
  - Probability of success is constant

Used for:
  - Modeling yes/no outcomes
  Example: Number of heads in 10 coin tosses
'''
```

```python
# What is the standard normal distribution, and why is it important?
'''
Standard normal distribution
  - A normal distribution with mean = 0
  - Standard deviation = 1
  - Symmetric, bell-shaped curve

Why it is important
  - Helps compute probabilities using z-scores
'''
```

```python
# What is the Central Limit Theorem (CLT), and why is it critical in statistics?
'''
Central Limit Theorem (CLT)
  - The distribution of sample means becomes normal as sample size increases
  - Applies regardless of population distribution

Why it is critical
  - Enables statistical inference
  - Justifies use of normal distribution for large samples
'''
```

```python
# What is the significance of confidence intervals in statistical analysis?
'''
Confidence intervals
  - Give a range where the true population parameter likely lies
  - Show uncertainty of an estimate
  - More informative than a single point estimate
'''
```

```python
# What is the concept of expected value in a probability distribution?
'''
Expected value
- The long-run average outcome of a random variable
- Calculated as Σ(value × probability)
- Represents the mean of the distribution
'''
```

```python
# Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution. (Include your Python code and output in the code box below.)

import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=50, scale=5, size=1000)

mean = np.mean(data)
std = np.std(data)

print("Mean:", mean)
print("Standard Deviation:", std)

plt.hist(data, bins=30)
plt.title("Normal Distribution (mean=50, std=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend. daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,235, 260, 245, 250, 225, 270, 265, 255, 250, 260] Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval. Write the Python code to compute the mean sales and its confidence interval.

import numpy as np
from math import sqrt

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

mean = np.mean(daily_sales)
std = np.std(daily_sales, ddof=1)
n = len(daily_sales)

z = 1.96  # 95% confidence
margin = z * (std / sqrt(n))

lower = mean - margin
upper = mean + margin

print("Mean sales:", mean)
print("95% Confidence Interval:", (lower, upper))
```
