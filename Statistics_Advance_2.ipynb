```python
# What is hypothesis testing in statistics?
'''
Hypothesis testing
  - A method to make decisions using sample data
  - Tests a claim about a population

Steps:
  - State Null hypothesis
  - State Alternative hypothesis
  - Calculate test statistic & p-value
  - Reject or fail to reject

Purpose:
  Check if an effect or difference is statistically significant
'''
```

```python
# What is the null hypothesis, and how does it differ from the alternative hypothesis?
'''
Null hypothesis (H0)
  - Assumes no effect / no difference
  - Default assumption
Alternative hypothesis (H1 or Ha)
  - Assumes there is an effect / difference
  - What we try to support
Difference:
  H0 = nothing changes
  H1 = something changes
'''
```

```python
#  Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
'''
Significance level (aplha)
  - The probability of rejecting a true null hypothesis
  - Common value: 0.05 (5%)
Role in decision:
  - If p-value ≤ aplha → Reject H0
  - If p-value > aplha → Fail to reject H0
Meaning:
  - aplha controls the risk of Type I error (false positive)
'''
```

```python
# What are Type I and Type II errors? Give examples of each.
'''
Type I Error (False Positive)
  Rejecting a true null hypothesis
  Example: Concluding a new drug works when it actually doesn’t
Type II Error (False Negative)
  Failing to reject a false null hypothesis
  Example: Concluding a new drug doesn’t work when it actually does
Simple:
  Type I → Detect effect that isn’t real
  Type II → Miss effect that is real
'''
```

```python
# What is the difference between a Z-test and a T-test? Explain when to use each.
'''
Z-test
  - Used when population standard deviation is known
  - Sample size is large (n ≥ 30)
  - Uses normal distribution
T-test
  - Used when population standard deviation is unknown
  - Sample size is small (n < 30)
  - Uses t-distribution
Simple rule:
  - Known σ → Z-test
  - Unknown σ → T-test
'''
```

```python
# Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.

import numpy as np
from math import sqrt
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu = 50
n = len(sample_data)
mean = np.mean(sample_data)
std = np.std(sample_data, ddof=1)
z = (mean - mu) / (std / sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Sample Mean:", mean)
print("Z-statistic:", z)
print("P-value:", p_value)
```

```python
# Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from math import sqrt

data = np.random.normal(loc=100, scale=15, size=1000)

mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

z = 1.96
margin = z * (std / sqrt(n))
lower = mean - margin
upper = mean + margin

print("Sample Mean:", mean)
print("95% Confidence Interval:", (lower, upper))

plt.hist(data, bins=30)
plt.title("Simulated Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = (data - mean) / std
    return z_scores

data = np.array([10, 12, 15, 18, 20, 22, 25, 30])

z_scores = calculate_z_scores(data)

print("Z-scores:", z_scores)

plt.hist(z_scores, bins=6)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
